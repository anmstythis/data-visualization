Представление первых семи данных о росте, весе и годе рождения персонажей в виде линейного графика

Для того, чтобы составить линейный график по выбранным данным, необходимо вызвать метод plot из библиотеки matplotlib.pyplot (которая была импортирована как plt). В диаграмму был установлен маркер в форме звёздочки размером 10 pt. Для роста был установлен синий цвет, для веса - оранжевый, для года - лаймовый. Для того, чтобы вывести получившийся график в терминал, необходимо вызвать метод show(). Метод legend() необязательный, однако если его не вызвать, то в диаграмму не выведутся названия графиков.

Примечание: в библиотеке matplotlib доступно множество цветов, но не все. Доступные цвета были найдены и изучены в официальной документации этой библиотеки.

```python
plt.figure(figsize=(10, 5))
plt.plot(char['name'].head(7), char['height'].head(7), marker='*', markersize=10, label="Рост", color='blue')
plt.plot(char['name'].head(7), char['mass'].head(7), marker='*', markersize=10, label="Вес", color='orange')
plt.plot(char['name'].head(7), char['birth_year'].head(7), marker='*', markersize=10, label="Год рождения", color='lime')
plt.title("Статистика среди персонажей \"Звездных войн\"")
plt.xlabel("Имена персонажей")
plt.ylabel("Числовые соотношения")
plt.legend()
plt.show()
```

Вывод: визуалированы данные о росте, весе и годе рождения первых семерых персонажей в виде линейного графика. Графики объединены в одну диаграмму.

Представление первых семи данных о росте, весе и годе рождения персонажей в виде столбчатой диаграммы

Берём абсолютно те же данные, но представляем их немного по-другому, т. е. через метод bar, а не plot. Параметр alpha был использован для слияния графиков. 

```python
plt.figure(figsize=(10, 5))
plt.bar(char['name'].head(7), char['height'].head(7), label="Рост", color='sienna')
plt.bar(char['name'].head(7), char['mass'].head(7), label="Вес", color='indigo', alpha= 0.5)
plt.bar(char['name'].head(7), char['birth_year'].head(7), label="Год рождения", color='darkolivegreen', alpha= 0.5)
plt.title("Статистика среди персонажей \"Звездных войн\"")
plt.xlabel("Имена персонажей")
plt.ylabel("Числовые соотношения")
plt.legend()
plt.show()
```

Вывод: визуалированы данные о росте, весе и годе рождения первых семерых персонажей в виде столбчатой диаграммы. Все диаграммы объединены в одну общую диаграмму.

Представление данных о поле персонажей в виде круговой диаграммы

Для того, чтобы получить количество данных, допустим с названием "male", мы сначала фильтруем данные в таблице, где находятся сведения о поле. Затем создаём списки со значением количества персонажей с мужским и женским полом, с учетом персонажей, у которых пол вовсе не указан. Для расчета количества данных был использован метод count(). Для создания круговой диаграммы нужно использовать метод pie(). Туда мы вносим список значений, сопоставляемых друг с другом, список названий этих значений, дополнительный параметр autopct, который демонстрирует значения в процентах, и список цветов каждого "кусочка". 

```python
male = char[char['sex'] == 'male']['sex']
female = char[char['sex'] == 'female']['sex']
nogend = char[char['sex'] == 'none']['sex']

gender = [male.count(), female.count(), nogend.count()]
labels = ["Мужчины", "Женщины", "Неизвестно"]


plt.pie(gender, labels= labels, autopct='%1.1f%%', colors= ['steelblue', 'lightpink', 'darkseagreen'])
plt.title("Соотношение пола персонажей \"Звёздных войн\"")
plt.show()
```

Вывод: визуалирована статистика с полом персонажей в виде круглой диаграммы.

Диаграмма по видам персонажей была выведена в терминал таким же образом, как и предыдущая

```python
humans = char[char['species'] == 'Human']['species']
droids = char[char['species'] == 'Droid']['species']

species = [humans.count(), droids.count()]
labels = ["Люди", "Дроиды"]

plt.pie(species, labels= labels, autopct='%1.1f%%', colors= ['moccasin', 'slategray'])
plt.title("Соотношение видов персонажей \"Звёздных войн\"")
plt.show()
```

Вывод: визуалирована статистика с видами персонажей в виде круглой диаграммы.

Представление данных в виде диаграммы рассеивания

Теперь приступаем к визуализации данных в виде диаграмм рассеивания. Для того, чтобы это реализовать, нужно использовать метод scatter(). Туда необходимо ввести данные, которые будут соотнесены между собой. Цвет вводить необязательно, но тогда диаграмма будет представлена рандомным цветом.

```python
plt.figure(figsize=(10, 6))
plt.scatter(char['height'], char['mass'], color='goldenrod')
plt.title("Соотношение между весом и ростом персонажей из \"Звёздных войн\"")
plt.xlabel("Рост")
plt.ylabel("Вес")
plt.show()
```

Вывод: визуалировано соотношение между весом и ростом персонажей в виде диаграммы рассеивания.

Вторая диаграмма была создана таким же образом, что и предыдущая.

```python
plt.figure(figsize=(6, 6))
plt.scatter(char['birth_year'], char['height'], color='cadetblue')
plt.title("Соотношение между годом рождения и ростом персонажей из \"Звёздных войн\"")
plt.xlabel("Год рождения")
plt.ylabel("Рост")
plt.show()
```

Вывод: визуалировано соотношение между год рождения и ростом персонажей в виде диаграммы рассеивания.

Представление данных в виде гистограммы

Для того, чтобы представить данные в виде гистограммы, необходимо вызвать метод hist(). Здесь рассматривается статистика какого-нибудь определённого столбца с данными, поэтому в начале объявляется этот столбец, а потом размер диапазона между значениями, основной цвет, цвет краев и название гистограммы. Дальше все то же самое, что и было в линейном графике.

```python
plt.figure(figsize=(10, 6))
plt.hist(char['birth_year'], bins=50, color='lavender', edgecolor='black', label='Распределение года рождения персонажей из \"Звёздных войн\"')
plt.legend()
plt.show()
```

Вывод: визуалирована статистика годов рождения персонажей в виде гистораммы.

```python
plt.figure(figsize=(10, 6))
plt.hist(char['mass'], bins=50, color='thistle', edgecolor='black', label='Распределение веса персонажей из \"Звёздных войн\"')
plt.legend()
plt.show()
```

Вывод: визуалирована статистика веса персонажей в виде гистораммы.

```python
plt.figure(figsize=(10, 6))
plt.hist(char['height'], bins=50, color='skyblue', edgecolor='black', label='Распределение роста персонажей из \"Звёздных войн\"')
plt.legend()
plt.show()
```

Вывод: визуалирована статистика роста персонажей в виде гистораммы.
